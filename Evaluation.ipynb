```python
from tqdm import tqdm
```

```python
import pandas as pd
from bert_score import score


df=dataset
# Initialize lists for BERTScore metrics
phi_scores, mistral_scores, deepseek_scores = [], [], []

# Compute BERTScore for each row
for idx, row in tqdm(df.iterrows()):
    # Reference and candidates
    ref = [row["article"]]

    # Calculate BERTScore for summary_phi
    P1, R1, F11 = score([row["Summary_Phi"]], ref, lang='en', verbose=False)
    phi_scores.append((P1.item(), R1.item(), F11.item()))

    # Calculate BERTScore for summary_mistral
    P2, R2, F12 = score([row["Summary_mistral"]], ref, lang='en', verbose=False)
    mistral_scores.append((P2.item(), R2.item(), F12.item()))

    # Calculate BERTScore for summary_deepseek
    P3, R3, F13 = score([row["summary_deepseek_qwen_new"]], ref, lang='en', verbose=False)
    deepseek_scores.append((P3.item(), R3.item(), F13.item()))

# Append BERTScores to DataFrame
df["P1"], df["R1"], df["F11"] = zip(*phi_scores)
df["P2"], df["R2"], df["F12"] = zip(*mistral_scores)
df["P3"], df["R3"], df["F13"] = zip(*deepseek_scores)
```

```python

# Reorder columns for clarity
df = df[[
    "article", "highlights",
    "Summary_Phi", "P1", "R1", "F11",
    "Summary_mistral", "P2", "R2", "F12",
    "summary_deepseek_qwen_new", "P3", "R3", "F13"
]]

# Display the resulting DataFrame
print(df)

# Save to CSV
df.to_csv("bert_score_results.csv", index=False)
print("Results saved to 'bert_score_results.csv'")

```

```python
import seaborn as sns
import numpy as np

# Creating a heatmap for precision, recall, and F1 scores
metrics_data = {
    'Model': ['Phi', 'Mistral', 'DeepSeek-Qwen'],
    'Precision': precision,
    'Recall': recall,
    'F1 Score': f1
}

metrics_df = pd.DataFrame(metrics_data).set_index('Model')

# Heatmap visualization
plt.figure(figsize=(8, 5))
sns.heatmap(metrics_df, annot=True, cmap='YlGnBu', linewidths=0.5)
plt.title('Heatmap of Precision, Recall, and F1 Score')
plt.show()

# Radar Chart for performance comparison
labels = np.array(['Precision', 'Recall', 'F1 Score'])
stats_phi = np.array([precision[0], recall[0], f1[0]])
stats_mistral = np.array([precision[1], recall[1], f1[1]])
stats_deepseek = np.array([precision[2], recall[2], f1[2]])

# Radar chart setup
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# Angles for the radar chart
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
angles += angles[:1]

# Plotting Phi
stats_phi = np.concatenate((stats_phi, [stats_phi[0]]))
ax.plot(angles, stats_phi, color='skyblue', linewidth=2, label='Phi')
ax.fill(angles, stats_phi, color='skyblue', alpha=0.3)

# Plotting Mistral
stats_mistral = np.concatenate((stats_mistral, [stats_mistral[0]]))
ax.plot(angles, stats_mistral, color='green', linewidth=2, label='Mistral')
ax.fill(angles, stats_mistral, color='green', alpha=0.3)

# Plotting DeepSeek-Qwen
stats_deepseek = np.concatenate((stats_deepseek, [stats_deepseek[0]]))
ax.plot(angles, stats_deepseek, color='salmon', linewidth=2, label='DeepSeek-Qwen')
ax.fill(angles, stats_deepseek, color='salmon', alpha=0.3)

# Add labels and title
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)
ax.set_title('Radar Chart of Model Performance')
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))

plt.show()
```
